```python
import os
import numpy as np
import torch
import matplotlib.pyplot as plt

import torchvision.transforms.functional as F


plt.rcParams["savefig.bbox"] = "tight"


```

```python
def show(imgs):
    if not isinstance(imgs, list):
        imgs = [imgs]
    fix, axs = plt.subplots(ncols=len(imgs), squeeze=False)
    for i, image in enumerate(imgs):
        image = image.detach()  # 转换为普通tensor，不携带梯度、设备等信息
        image = F.to_pil_image(image)  # 转换为PIL Image
        axs[0, i].imshow(np.asarray(image))
        axs[0, i].set(xticklabels=[], yticklabels=[], xticks=[], yticks=[])
```

```python
import json
from PIL import Image

with open("img_mask_path.json", "r") as f:
    img_mask_path = json.load(f)

img_path = r"G:\Program\DATABASE_JPG\2020\LanYanLing\LanYanLing_70.jpg".replace("\\", "/")
# print(img_mask_path[img_path])
img = Image.open(img_path).convert("RGB")
mask1 = Image.open(img_mask_path[img_path][0])
mask2 = Image.open(img_mask_path[img_path][1])

img = F.to_tensor(img)
mask1 = F.to_tensor(mask1)
mask2 = F.to_tensor(mask2)
```

```python
show([img, mask1, mask2])
```
